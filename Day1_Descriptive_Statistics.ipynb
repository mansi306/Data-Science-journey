Analyzing a 7-day sales profit dataset to understand the impact of outliers (like a sudden 15,000 spike) on the Mean and Median.


> Add blockquote



```python
import numpy as np
sales_profit = [2000, 2500, 2000, 3000, 15000, 2200, 2100]
mean_profit = np.mean(sales_profit)
median_profit = np.median(sales_profit)
print("Mean Profit : ",mean_profit)
print("Median_profit : ",median_profit)
```
